## 1. Let's extract the prompts that have left and right in them and create some datasets.

```python
left_and_right_relations = ["left", "right"]
lr_dataset_df = dataset_df[dataset_df['output_text'].str.contains('|'.join(left_and_right_relations), case=False)]

# Reset the index of the filtered DataFrame
lr_dataset_df = lr_dataset_df.reset_index(drop=True)
```

```python
lr_dataset_df
```

```python
len(lr_dataset_df)
```

```python
lr_dataset_df.to_csv('/u/vbutoi/victor_space/data/LocalizedNarratives/LeftRight_LN_HF_Dataset.csv', index=False)
```

### We can do another version where we have the relations and objects listed.

```python
lr_dataset_df_v2 = lr_dataset_df.copy()
```

```python
def lr_row_operation(row):
    options = "Objects: " + row['input_text'] + ". "
    target_relations = ['left', 'right']
    found = []
    relations = "Relations: "
    for rel in target_relations:
        if rel in row["output_text"]:
            found.append(rel)
    relations += ", ".join(found)
    return options + relations

lr_dataset_df_v2['input_text'] = lr_dataset_df.apply(lr_row_operation, axis=1)
```

```python
lr_dataset_df_v2
```

```python
lr_dataset_df_v2.to_csv('/u/vbutoi/victor_space/data/LocalizedNarratives/LeftRight_v2_LN_HF_Dataset.csv', index=False)
```
